```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Sample user-item rating matrix
data = {
    'User1': [5, 3, 0, 1],
    'User2': [4, 0, 0, 1],
    'User3': [1, 1, 0, 5],
    'User4': [1, 0, 0, 4],
    'User5': [0, 1, 5, 4],
}
df = pd.DataFrame(data, index=['ItemA', 'ItemB', 'ItemC', 'ItemD'])
df
```
